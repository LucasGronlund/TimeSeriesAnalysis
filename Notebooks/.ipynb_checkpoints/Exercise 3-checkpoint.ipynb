```python
#imports
import numpy as np
import pandas as pd
import pylab as pb
import statsmodels.tsa.api as smt
from statsmodels.tsa.api import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace import sarimax
```

```python
#load data
arctic = pd.read_csv('Exercises_data/arctic.csv')
```

```python
decomp = smt.seasonal_decompose(arctic['x'].as_matrix(),freq=12,model='additive',two_sided=False)
fig = decomp.plot()
fig.set_size_inches(15,10)
pb.show()
print("\nLooks like there is a slightly decreasing trend and a clear seasonal component with period 12.")
```

```python
pb.plot(decomp.resid)
pb.show()
print("dickey fuller gives p-value: " + repr(adfuller(decomp.resid[6:len(decomp.resid)-6])[1]))
```

```python
seasonal_diff = arctic.diff(12)
seasonal_diff.plot(title="seasonal component removed")
pb.show()
print("dickey-fuller test gives p-value: " + repr(adfuller(seasonal_diff['x'].as_matrix()[12:])[1]),end="\n\n")


true_diff = seasonal_diff.diff(1)
true_diff.plot(title="differed once")
pb.show()
true_diff.drop(np.arange(0,13),inplace=True)
print("dickey-fuller test gives p-value: " + repr(adfuller(true_diff['x'])[1]))

testShit.plot(ax=ax)
pb.show()
```
