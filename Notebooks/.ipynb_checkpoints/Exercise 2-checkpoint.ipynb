# Exercise 2 in time series analysis

```python
#Imports 
import numpy as np
import pylab as pb
from statsmodels.tsa.stattools import acf, pacf as acf2, pacf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
import pandas as pd
from pandas.plotting import autocorrelation_plot as acf
```

```python
#methods
def model(coeff,nrData,X_t = 0):
    #return as [X_1,X_2,...,X_t]
    ts = np.zeros(nrData);
    noise = np.zeros(nrData);
    for i in range(nrData):
        Z_t = np.random.normal(0,1);
        X_t = coeff*X_t + Z_t;
        ts[i] = X_t;
        noise[i] = Z_t;
    return ts, noise

def autocorr(x):
    #assumed model goes from -inf -> inf, so we
    #cut at half cuz we want h from 0 -> "inf"
    result = np.correlate(x, x, mode='full')
    return result[result.size//2:]/np.max(result)
```

```python
pb.close()
nrDatapoints = 100;


phi = 0.5;
series = model(phi,nrDatapoints,np.random.normal(0,2/np.sqrt(3)));
fig, ax = pb.subplots(2,2,figsize=(15,10))
ax[0,0].plot(np.arange(nrDatapoints),series[0])
ax[0,0].set_title("time series")
ax[0,1].plot(np.arange(nrDatapoints),series[1])
ax[0,1].set_title("noise")
acf(pd.Series(series[0]),ax=ax[1,0]);
ax[1,0].set_title("actual acf")
ax[1,1].plot(np.arange(nrDatapoints),0.5**np.arange(nrDatapoints))
ax[1,1].set_title("analytical acf")
fig.suptitle("using phi = 1.05")
pb.show()


pb.show()
```
