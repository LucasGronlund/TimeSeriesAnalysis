# Exercise 1 in time series analysis

```python
#Imports 
import numpy as np
import pylab as pb
```

```python
#methods
def model(coeff,nrData):
    ts = np.zeros(nrData);
    noise = np.zeros(nrData);
    X_t = 0; #initial data = 0
    for i in range(nrData):
        Z_t = np.random.normal(0,1);
        X_t = coeff*X_t + Z_t;
        ts[i] = X_t;
        noise[i] = Z_t;
    return ts, noise
```

```python
pb.close()
nrDatapoints = 50;


phi = 1.05;
series = model(phi,nrDatapoints);
fig, ax = pb.subplots(1,2,figsize=(15,5))
ax[0].plot(np.arange(nrDatapoints),series[0])
ax[0].set_title("time series")
ax[1].plot(np.arange(nrDatapoints),series[1])
ax[1].set_title("noise")
fig.suptitle("using phi = 1.05")
pb.show()


phi = 0.95;
series = model(phi,nrDatapoints);
fig, ax = pb.subplots(1,2,figsize=(15,5))
ax[0].plot(np.arange(nrDatapoints),series[0])
ax[0].set_title("time series")
ax[1].plot(np.arange(nrDatapoints),series[1])
ax[1].set_title("noise")
fig.suptitle("using phi = 0.95")
pb.show()
```
