```python
#Imports
import numpy as np
import pandas as pd
import pylab as pb
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsmodels.tsa.stattools import acf, pacf as acf, pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.gofplots import qqplot
from statsmodels.tsa.arima_model import ARMA, ARIMA
import statsmodels.tsa.api as smt
from scipy.optimize import leastsq
from scipy.stats import norm
from statsmodels.sandbox.tsa.garch import Garch, garchplot, Garch0, optimize
from arch import arch_model
import statsmodels.sandbox.tsa.fftarma as fftarma
from statsmodels.tsa.stattools import periodogram
from numpy import pi
```

```python
#Example
# GARCH(1,1) Model in Python
#   uses maximum likelihood method to estimate (omega,alpha,beta)
# (c) 2014 QuantAtRisk, by Pawel Lachowicz; tested with Python 3.5 only
 
r = np.array([0.945532630498276,
              0.614772790142383,
              0.834417758890680,
              0.862344782601800,
              0.555858715401929,
              0.641058419842652,
              0.720118656981704,
              0.643948007732270,
              0.138790608092353,
              0.279264178231250,
              0.993836948076485,
              0.531967023876420,
              0.964455754192395,
              0.873171802181126,
              0.937828816793698])
 
from arch import arch_model
garch11 = arch_model(r, p=1, q=1)
res = garch11.fit(update_freq=1)
print(res.summary())
res.plot()
pb.show()
```

```python
print("\t\t\t\t\t\t-----White noise-----")
nrData = 1000;
fig, ax = pb.subplots(2,1,figsize=(15,10))
wn = np.random.rand(nrData)-0.5
ax[0].plot(wn)

ax[1].plot(periodogram(wn))
pb.show()
```

```python
print("\t\t\t\t\t\t-----Cosine-----")
nrData = 100;
data = np.linspace(0,nrData,nrData)
value = np.cos(data);
fig, ax = pb.subplots(2,1,figsize=(15,10))
ax[0].plot(data,value)

ax[1].plot(data,periodogram(value)/(2*pi))
pb.show()
```

```python
from scipy.signal import periodogram as pgram
pdData = pd.read_csv('Exercises_data/porkprices.txt',
                     sep=' ',header=None);
pdData = pdData.transpose()
data = pdData.as_matrix()

pdData.plot()
pb.show()

diffedData = pdData[0].diff(12)

diffedData.plot()
pb.show()

period = periodogram(diffedData.as_matrix()[12:]);
pb.plot(period)
pb.xlim([0,30])
pb.show()

print("\t-----peak at 5-----")
print("period is: " + repr(5/len(diffedData)*2*pi))
```

```python
from scipy.signal import periodogram as pgram
pdData = pd.read_csv('Exercises_data/porkprices.txt',
                     sep=' ',header=None);
pdData = pdData.transpose()
data = pdData.as_matrix()

pdData.plot()
pb.show()

diffedData = pdData[0].diff(12)

diffedData.plot()
pb.show()

asd, period = pgram(diffedData.as_matrix()[12:]);
pb.plot(period)
pb.xlim(0,30)
pb.show()

print("\t-----peak at 5-----")
#Maybe think of the value 5 aas the number of peaks in this case. Then we can think of
#the following maths as just dividing the number of peaks with the length of the data to get
#the period. Then transform to frequency by forumula: period*freq=2pi
print("period is: " + repr(5/(len(diffedData))*2*pi))
```
